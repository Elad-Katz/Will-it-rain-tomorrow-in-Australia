```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/Raanank10/Rain_in_Australia/main/weatherAUS.csv'

data = pd.read_csv(url)
data.head(5)
```

```python
Location_list = pd.DataFrame(data.Location.unique())
Location_list.rename(columns = {0:'Location'},inplace=True)
Location_list.head()
```

```python
type(Location_list)
```

```python
enc = OrdinalEncoder()
```

```python
# enc.fit(df[["Sex","Blood", "Study"]])
enc.fit(Location_list[['Location']])
```

```python
Location_list.head()
```

```python
Location_list['Location_as_code'] = enc.transform(Location_list[["Location"]])
#df[["Sex","Blood", "Study"]] = enc.transform(df[["Sex","Blood", "Study"]])
```
